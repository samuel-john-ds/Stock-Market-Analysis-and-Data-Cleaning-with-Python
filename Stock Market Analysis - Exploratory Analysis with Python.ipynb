```python
sector_count = sp500_companies_sym_df['Sector'].value_counts()
```

```python
# Plot a pie chart from the distribution of company sectors

sector_count.plot(kind='pie', figsize = (10,10))
```

```python
# Plot a bar chart from the distribution of company sectors

sector_count.plot(kind='bar', figsize = (10,10), xlabel='Sector',ylabel='Number of Companies')
```

# Find companies that falls under Sector > Consumer Discretionary

```python
sp500_companies_sym_df[sp500_companies_sym_df['Sector'] == 'Consumer Discretionary']
```

# Sort by Price/Earnings in ascending order

```python
sp500_companies_sym_df.sort_values('Price/Earnings',ascending= False)
```

___

# Find Companies in Information technology and sort from Highest Price/Earnings to Lowest

```python
sp500_companies_sym_df[sp500_companies_sym_df['Sector']=='Information Technology'].sort_values('Price/Earnings',ascending = False)
```

```python
sp500_companies_sym_df.head()
```

```python
sector_marketcap_df = sp500_companies_sym_df[['Sector','Market Cap']]
```

```python
sector_marketcap_df.head()
```

```python
sector_marketcap_df.sort_values(['Sector','Market Cap'], ascending=[False,False])
```

```python
sector_marketcap_df.value_counts(sort=True)
```

# Find Sectors and sort from Highest Market Cap to Lowest

```python
sector_marketcap_df.groupby("Sector")["Market Cap"].sum().sort_values(ascending=False)
```

```python
sector_marketcap_total_df = sector_marketcap_df.groupby("Sector")["Market Cap"].sum().sort_values(ascending=False)
```

```python
sector_marketcap_total_df.plot(kind='bar', x='Sector', y='Market Cap', ylabel = 'Market Cap', title='Market Cap in each Sector',figsize=(10,10))
```

```python
sector_marketcap_df[sector_marketcap_df['Sector']=='Information Technology'].sort_values("Market Cap",ascending = False)
```

```python
sp500_companies_sym_df[['Name','Sector','Market Cap']].sort_values('Market Cap',ascending = False).head(10)
```

___
